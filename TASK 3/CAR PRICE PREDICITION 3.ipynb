```python
predict="price"
d=d[["symboling","wheelbase","carlength","carwidth","carheight","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]]
x=np.array(d.drop([predict],1))
y=np.array(d[predict])
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
from sklearn.metrics import mean_absolute_error
model.score(x_test,predictions)
```
